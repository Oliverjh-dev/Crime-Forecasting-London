1. Purpose

Train the forecasting models using the cleaned and feature-engineered dataset.

2. Modelling Approach

Two statistical modelling approaches are used:

A. Negative Binomial GLM

Suitable for overdispersed crime count data

Includes seasonal features and IMD score

Boroughs encoded as fixed effects

B. Hierarchical Mixed-Effects Model

Borough-specific random intercepts

Captures structural differences

Adds interpretability

Why not pure machine learning?

This project is designed to:

Demonstrate statistical modelling strength

Provide interpretable results

Fit a real-world crime forecasting problem

3. Model Evaluation
